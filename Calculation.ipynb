```python
import numpy as np
import pandas as pd

pd.set_option("display.max_columns", None)
```

# Load Data

```python
df = pd.read_parquet("skor-abm-detail.parquet")

for col in df.columns:
    if "minat" in col:
        df[f"{col}_ket"] = df[col].apply(lambda x: "Minat" if x >= 60 else "Tidak Minat")

df.head()
```

```python
df_report = pd.read_excel("dim_abm_report.xlsx", sheet_name="Minat")
df_report
```

# Melt Data

In this step, we are restructuring the DataFrame to make it easier to analyze and work with data related to interest levels. The `pd.melt` function is used to transform the data from a **wide format**, where information is spread across multiple columns, to a **long format**, where the same information is consolidated into two new columns.

#### Breakdown of the Code:
1. **`id_vars=["nm_prop"]`:** This specifies that the `nm_prop` column remains unchanged. It acts as an identifier, ensuring that its values are preserved in the resulting DataFrame.
2. **`value_vars`:** Here, we dynamically select all columns that contain the substrings "minat" and "ket" in their names. These columns likely represent interest-related attributes across different categories.
3. **`var_name="minat"`:** The names of the selected columns (e.g., `minat_1_ket`, `minat_2_ket`, etc.) are moved into a new column named `minat`.
4. **`value_name="minat_level"`:** The data from the selected columns is moved into a new column named `minat_level`, which holds the corresponding values.

#### Example Before Transformation:
| nm_prop  | minat_1_ket | minat_2_ket | minat_3_ket |  
|----------|-------------|-------------|-------------|  
| Prop A   | High        | Low         | Medium      |  
| Prop B   | Low         | High        | Medium      |  

#### Example After Transformation:
| nm_prop  | minat        | minat_level |  
|----------|--------------|-------------|  
| Prop A   | minat_1_ket  | High        |  
| Prop A   | minat_2_ket  | Low         |  
| Prop A   | minat_3_ket  | Medium      |  
| Prop B   | minat_1_ket  | Low         |  
| Prop B   | minat_2_ket  | High        |  
| Prop B   | minat_3_ket  | Medium      | 

```python
df_melted = pd.melt(
    frame=df,
    id_vars=["nm_prop"],
    value_vars=[col for col in df.columns if ("minat" in col) and ("ket" in col)],
    var_name="minat",
    value_name="minat_level"
)
df_melted
```

In this step, we further clean and enhance the transformed DataFrame by modifying the `minat` column to make it more meaningful. This involves two main operations: removing unnecessary substrings and mapping descriptive labels to the values in the `minat` column.

#### Breakdown of the Code:
1. **`df_melted.copy()`:** A copy of the `df_melted` DataFrame is made to ensure the original data remains unchanged.
2. **Removing `"_ket"` from `minat`:** The `str.replace("_ket", "")` operation removes the "_ket" substring from the values in the `minat` column. This simplifies the column names by removing redundant parts, leaving only the relevant identifiers.
3. **Mapping Descriptive Labels:**  
   - The `map` function is used to replace the values in the `minat` column with more descriptive labels.  
   - The mapping is created using `df_report.set_index("Aspek_id")["Aspek_desc"].to_dict()`, which generates a dictionary where the keys are `Aspek_id` values (corresponding to the original `minat` identifiers) and the values are `Aspek_desc` (the descriptive labels).  
   - Each value in the `minat` column is then replaced with its corresponding descriptive label using this dictionary.

```python
df_report.set_index("Aspek_id")["Aspek_desc"].to_dict()
```

```python
df_melted_cleaned = df_melted.copy()
df_melted_cleaned["minat"] = df_melted_cleaned["minat"].str.replace("_ket", "")
df_melted_cleaned["minat"] = df_melted_cleaned["minat"].map(df_report.set_index("Aspek_id")["Aspek_desc"].to_dict())
df_melted_cleaned
```

# Group Data

In this step, we group the cleaned data to summarize the frequency of each combination of `nm_prop`, `minat`, and `minat_level`. This provides insights into how often specific interest levels appear for each category.

#### Breakdown of the Code:
1. **`df_melted_cleaned.copy()`:** A copy of the `df_melted_cleaned` DataFrame is made to ensure the original data remains unchanged.
2. **Grouping with `groupby`:**  
   - The `groupby` method is used to group the data by the columns `nm_prop`, `minat`, and `minat_level`.  
   - This operation creates groups for each unique combination of these columns.
3. **Calculating the Group Size:**  
   - The `size()` function is applied to count the number of rows in each group (i.e., the frequency of each combination).  
4. **Resetting the Index:**  
   - The `reset_index()` method is used to convert the grouped data back into a standard DataFrame.  
   - The resulting DataFrame has the original grouping columns (`nm_prop`, `minat`, and `minat_level`) along with a new column indicating the count of rows in each group.

#### Example Before Grouping:
| nm_prop  | minat            | minat_level |  
|----------|------------------|-------------|  
| Prop A   | Interest Level 1 | High        |  
| Prop A   | Interest Level 1 | High        |  
| Prop A   | Interest Level 2 | Low         |  

#### Example After Grouping:
| nm_prop  | minat            | minat_level | count |  
|----------|------------------|-------------|-------|  
| Prop A   | Interest Level 1 | High        | 2     |  
| Prop A   | Interest Level 2 | Low         | 1     |  

```python
df_grouped = df_melted_cleaned.copy()
df_grouped = df_grouped.groupby(["nm_prop", "minat", "minat_level"]).size().reset_index()
df_grouped
```

In this step, we further clean the grouped data by renaming the columns for clarity and adding a new column that categorizes the interest levels based on the description.

#### Breakdown of the Code:
1. **Renaming Columns:**
   - The `columns` attribute is used to rename the columns of the grouped DataFrame. The new column names are:
     - `province` (previously `nm_prop`) â€“ representing the region.
     - `minat_desc` (previously `minat`) â€“ representing the description of the interest level.
     - `minat_level` â€“ representing the actual interest level.
     - `count` â€“ representing the frequency of each combination.
2. **Mapping `minat_category`:**
   - The `map` function is applied to the `minat_desc` column to add a new column `minat_category`.  
   - This column is populated by mapping the `minat_desc` values to their corresponding categories from the `df_report` DataFrame. The mapping is created using the `Aspek_desc` column as keys and the `Kategori` column as values.
3. **Reordering Columns:**
   - We reorder the columns to display `province`, `minat_category`, `minat_desc`, `minat_level`, and `count` in this specific order to make the DataFrame easier to read and interpret.

```python
df_report.set_index("Aspek_desc")["Kategori"].to_dict()
```

```python
df_grouped_cleaned = df_grouped.copy()
df_grouped_cleaned.columns = ["province", "minat_desc", "minat_level", "count"]
df_grouped_cleaned["minat_category"] = df_grouped_cleaned["minat_desc"].map(df_report.set_index("Aspek_desc")["Kategori"].to_dict())
df_grouped_cleaned = df_grouped_cleaned[["province", "minat_category", "minat_desc", "minat_level", "count"]]
df_grouped_cleaned
```

# Pivot Data

In this step, we reshape the data once again to present it in a more compact and readable format using the `pivot` function. This allows us to organize the counts of interest levels across different categories into separate columns.

#### Breakdown of the Code:
1. **Creating a Copy of the DataFrame:**  
   - A copy of the `df_grouped_cleaned` DataFrame is made to ensure the original data remains unchanged.
   
2. **Pivoting the DataFrame:**  
   - The `pivot` function is applied to reshape the data. The `pivot` function rearranges the data such that:
     - **`index=["province", "minat_category", "minat_desc"]`:** These columns are set as the index (row labels) of the new DataFrame. This means each row will represent a unique combination of province, interest category, and interest description.
     - **`columns="minat_level"`:** The values in the `minat_level` column are spread across new columns, creating separate columns for each unique interest level.
     - **`values="count"`:** The `count` column is used as the data that fills the new DataFrame, representing the frequency of each interest level for each combination of `province`, `minat_category`, and `minat_desc`.

#### Example Before Pivoting:
| province | minat_category | minat_desc     | minat_level | count |  
|----------|----------------|----------------|-------------|-------|  
| Prop A   | Category 1     | Interest Level 1 | High        | 2     |  
| Prop A   | Category 1     | Interest Level 1 | Low         | 1     |  
| Prop B   | Category 2     | Interest Level 2 | High        | 3     |  

#### Example After Pivoting:
| province | minat_category | minat_desc     | High | Low |  
|----------|----------------|----------------|------|-----|  
| Prop A   | Category 1     | Interest Level 1 | 2    | 1   |  
| Prop B   | Category 2     | Interest Level 2 | 3    | 0   |  

```python
df_pivoted = df_grouped_cleaned.copy()
df_pivoted = df_pivoted.pivot(
    index=["province", "minat_category", "minat_desc"],
    columns="minat_level",
    values="count"
)
df_pivoted
```

In this step, we refine the pivoted DataFrame to make it cleaner, more interpretable, and analysis-ready. This involves resetting the index, renaming columns, sorting rows, and calculating percentages.

#### Breakdown of the Code:
1. **Creating a Copy:**  
   - A copy of the `df_pivoted` DataFrame is made to ensure the original data remains unchanged.

2. **Renaming Columns:**  
   - The column names are cleaned and renamed:
     - If the column name is a tuple (from the pivot operation), `"_".join(map(str, col))` combines the elements of the tuple into a single string.  
     - All columns are renamed to user-friendly names for clarity:
       - `Provinsi` â€“ Province name.
       - `Kategori Minat` â€“ Interest category.
       - `Deskripsi Minat` â€“ Interest description.
       - `Minat (Count)` â€“ Count of "Interested" responses.
       - `Tidak Minat (Count)` â€“ Count of "Not Interested" responses.

3. **Sorting the Data:**  
   - The DataFrame is sorted by:
     - `Provinsi` (alphabetically).
     - `Kategori Minat` (descending).
     - `Minat (Count)` (descending).  
   - Sorting ensures the data is organized for easier interpretation and analysis.

```python
df_pivoted_cleaned = df_pivoted.copy()
df_pivoted_cleaned = df_pivoted_cleaned.reset_index()
df_pivoted_cleaned.columns = ["_".join(map(str, col)).strip() if isinstance(col, tuple) else col for col in df_pivoted_cleaned.columns]
df_pivoted_cleaned.columns = ["Provinsi", "Kategori Minat", "Deskripsi Minat", "Minat (Count)", "Tidak Minat (Count)"]
df_pivoted_cleaned = df_pivoted_cleaned.sort_values(
    by=["Provinsi", "Kategori Minat", "Minat (Count)"],
    ascending=[True, False, False],
    ignore_index=True
)
df_pivoted_cleaned["Minat (Pct)"] = df_pivoted_cleaned["Minat (Count)"] / (df_pivoted_cleaned["Minat (Count)"] + df_pivoted_cleaned["Tidak Minat (Count)"])
df_pivoted_cleaned["Tidak Minat (Pct)"] = 1 - df_pivoted_cleaned["Minat (Pct)"]
df_pivoted_cleaned
```
