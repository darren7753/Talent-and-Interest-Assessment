```python
import numpy as np
import pandas as pd

pd.set_option("display.max_columns", None)
```

# Load Data

```python
df = pd.read_parquet("skor-abm-detail.parquet")

for col in df.columns:
    if "minat" in col:
        df[f"{col}_ket"] = df[col].apply(lambda x: "Minat" if x >= 60 else "Tidak Minat")

df.head()
```

```python
df_report = pd.read_excel("dim_abm_report.xlsx", sheet_name="Minat")
df_report
```

# Melt Data

```python
df_melted = pd.melt(
    frame=df,
    id_vars=["nm_prop"],
    value_vars=[col for col in df.columns if ("minat" in col) and ("ket" in col)],
    var_name="minat",
    value_name="minat_level"
)
df_melted
```

```python
df_report.set_index("Aspek_id")["Aspek_desc"].to_dict()
```

```python
df_melted_cleaned = df_melted.copy()
df_melted_cleaned["minat"] = df_melted_cleaned["minat"].str.replace("_ket", "")
df_melted_cleaned["minat"] = df_melted_cleaned["minat"].map(df_report.set_index("Aspek_id")["Aspek_desc"].to_dict())
df_melted_cleaned
```

# Group Data

```python
df_grouped = df_melted_cleaned.copy()
df_grouped = df_grouped.groupby(["nm_prop", "minat", "minat_level"]).size().reset_index()
df_grouped
```

```python
df_report.set_index("Aspek_desc")["Kategori"].to_dict()
```

```python
df_grouped_cleaned = df_grouped.copy()
df_grouped_cleaned.columns = ["province", "minat_desc", "minat_level", "count"]
df_grouped_cleaned["pct"] = df_grouped_cleaned["count"] / df_grouped_cleaned.groupby(["province", "minat_desc"])["count"].transform("sum")
df_grouped_cleaned["minat_category"] = df_grouped_cleaned["minat_desc"].map(df_report.set_index("Aspek_desc")["Kategori"].to_dict())
df_grouped_cleaned = df_grouped_cleaned[["province", "minat_category", "minat_desc", "minat_level", "count", "pct"]]
df_grouped_cleaned
```

# Pivot Data

```python
df_pivoted = df_grouped_cleaned.copy()
df_pivoted = df_pivoted.pivot(
    index=["province", "minat_category", "minat_desc"],
    columns="minat_level",
    values="pct"
)
df_pivoted
```

```python
df_pivoted_cleaned = df_pivoted.copy()
df_pivoted_cleaned = df_pivoted_cleaned.reset_index()
df_pivoted_cleaned = df_pivoted_cleaned.rename_axis(None, axis=1)
df_pivoted_cleaned = df_pivoted_cleaned.sort_values(
    by=["province", "minat_category", "Minat"],
    ascending=[True, False, False],
    ignore_index=True
)
df_pivoted_cleaned
```
